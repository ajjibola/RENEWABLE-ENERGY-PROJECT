Evaluating the correlation of renewable energy potentials between continents.

```python

correlation = {'Africa': [132.840769,9.029089,14.685940,8.161023],
        'Europe': [661.562560,139.052063,404.369480,217.632497],
        'Australia': [19.634,12.0811,16.2665,4.275],
               'North America': [733.477832,102.907233,322.528364,100.353372],
               'Asia Pacific': [1718.508325,314.208553,460.469456,221.301517]
        }

df = pd.DataFrame(correlation,columns=['Africa','Europe','Australia','North America', 'Asia Pacific'])
print (df)
```

```python
corrMatrix = df.corr()
print (corrMatrix)
```

```python
import seaborn as sn
sn.heatmap(corrMatrix, annot=True)
plt.show()
```
